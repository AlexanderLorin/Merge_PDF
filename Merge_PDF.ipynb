```python
path3 = "C:/Users/alorin/Documents/code/Billing/Merged"
```

```python

f1_files = os.listdir(path1)
f2_files = os.listdir(path2)

def pdf_merger(f1,f2):
    merger = PdfFileMerger()
    f1_content = PdfFileReader(open(os.path.join(path1,f1), 'rb'))
    f2_content = PdfFileReader(open(os.path.join(path2,f2), 'rb'))
    merger.append(f1_content)
    merger.append(f2_content)
    out = os.path.join(path3,f"{f1}")
    merger.write(out)
    
```

```python
for file1 in f1_files : 
    for file2 in f2_files: 
        if file1 in file2: 
            pdf_merger(file1,file2)
```
